```python
#zad1
import numpy as np
import matplotlib.pyplot as plt

def z1(x):
    return (6-2*x)/(6+4*x+x**2)

def z2(x):
    return (6-4*x+x**2)/(6+2*x)

x = np.linspace(-8, 8, 300)

wartości_1z= z1(x)
wartości_2z = z2(x)

plt.figure(figsize=(10, 6))
plt.plot(x, wartości_1z, label="przybliżenie z1", color="darkcyan", linewidth=1.3)
plt.plot(x, wartości_2z, label="przybliżenie z2", color="plum", linewidth=1.3)
plt.title("Przybliżenie Padégo funkcji exp(-x)", fontsize=12, fontweight='bold', color="dimgray")
plt.xlabel("x",fontsize=10, fontweight='bold', color="dimgray")
plt.ylabel("wartość przybliżenia",fontsize=10, fontweight='bold', color="dimgray")
plt.gca().set_facecolor('snow')
plt.legend()
plt.grid(True)

plt.savefig('wykres_przybliżenie_padego.png', format='png', dpi=300)

plt.show()



```

```python
fig, axs = plt.subplots(2, 1, figsize=(10, 10)) 

axs[0].plot(x, wartości_1z, label="z1 approximation", color="mediumvioletred", linewidth=1.5)
axs[0].set_title("Przybliżenie Padégo dla z1",fontsize=10, fontweight='bold', color="dimgray")
axs[0].set_xlabel("x")
axs[0].set_ylabel("wartość przybliżenia")
axs[0].axhline(0, color='black', linewidth=0.5)
axs[0].axvline(0, color='black', linewidth=0.5)
axs[0].grid(True)

axs[1].plot(x, wartości_2z, label="z2 approximation", color="olivedrab", linewidth=1.5)
axs[1].set_title("Przybliżenie Padégo dla z2",fontsize=10, fontweight='bold', color="dimgray")
axs[1].set_xlabel("x")
axs[1].set_ylabel("wartość przybliżenia")
axs[1].axhline(0, color='black', linewidth=0.5)
axs[1].axvline(0, color='black', linewidth=0.5)
axs[1].grid(True)


for ax in axs:
    ax.set_facecolor('snow')

plt.tight_layout()
plt.show()
```

```python
#zad2

def dokladnosc_maszynowa():
    epsilon = 1.0
    while (1.0 + epsilon) != 1.0:  
        epsilon /= 2.0             
    return epsilon * 2.0           

dokladnosc = dokladnosc_maszynowa()
print(f"Dokładność maszynowa: {dokladnosc}")
```
