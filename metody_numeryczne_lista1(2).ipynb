```python
#zad1
import numpy as np
import matplotlib.pyplot as plt

def z1(x):
    return (6-2*x)/(6+4*x+x**2)

def z2(x):
    return (6-4*x+x**2)/(6+2*x)

def funkcja(x):
    return np.exp(-x)

x = np.linspace(-8, 8, 300)

wartoci_1z= z1(x)
wartoci_2z = z2(x)
wartoci_funkcji=funkcja(x)

plt.figure(figsize=(10, 6))
plt.plot(x, wartoci_1z, label="przyblienie z1", color="darkcyan", linewidth=1.3)
plt.plot(x, wartoci_2z, label="przyblienie z2", color="plum", linewidth=1.3)
plt.plot(x, wartoci_funkcji, label="funkcja exp(-x)", color="gold", linewidth=1.3)
plt.title("Przyblienie Padgo funkcji exp(-x)", fontsize=12, fontweight='bold', color="dimgray")
plt.xlabel("x",fontsize=10, fontweight='bold', color="dimgray")
plt.ylabel("warto przyblienia",fontsize=10, fontweight='bold', color="dimgray")
plt.gca().set_facecolor('snow')
plt.legend()
plt.grid(True)

plt.savefig('/Users/zuzia/Desktop/wykres_przyblienie_padego.png', format='png', dpi=300)

plt.show()



```

```python
#zad2
def dokladnosc_maszynowa():
    epsilon = 1.0
    while (1.0 + epsilon) != 1.0:  
        epsilon /= 2.0             
    return epsilon * 2.0           

dokladnosc = dokladnosc_maszynowa()
print(f"Dokadno maszynowa: {dokladnosc}")
```

Notatka do zadania 2:
Szukamy dokadno maszynow komputera, ktra jest najmniejsz wartoci epsilon speniajc warunek 1 +  = 1. Stosujemy ptl i wraz z kad iteracj dzielimy epsilon przez 2 i ponownie sprawdzamy warunek. Ptla zatrzymuje si, gdy 1.0 +  nie jest ju wiksze od 1.0 (w arytmetyce zmiennoprzecinkowej komputer zaokrgla t warto do 1.0). Ostatnia warto  ktra jeszcze speniaa warunek 1 +  > 1.0, bya dwukrotnie wiksza ni ta, przy ktrej ptla si zatrzymaa, dlatego na koniec zwracamy warto epsilon razy dwa.

```python
#zad5
import numpy as np 
import time  

def W(x):
    return 6*x**4 + 5*x**3 - 13*x**2 + x + 1

def obliczenia():
    start_time = time.time()  
    x_wartosci = np.arange(-10, 10, 0.0001)
    w_wartosci = [W(x) for x in x_wartosci]

    with open("wielomian_wyniki.txt", "w") as file:
        for x, w in zip(x_wartosci, w_wartosci):
            file.write(f'W({x:.4f}) = {w:.4f}\n')
    
    end_time = time.time() 
    print(f"Czas wykonania: {end_time - start_time:.4f} sekundy") 

obliczenia()
```

```python
import numpy as np 
import time

def W_horner(x_wartosci):
    a = np.array([6, 5, -13, 1, 1])
    result = a[0] * np.ones_like(x_wartosci) 
    for coefficient in a[1:]:
        result = result * x_wartosci + coefficient
    return result

def obliczenia_horner():
    start_time = time.time()
    x_wartosci = np.arange(-10, 10, 0.0001)
    w_wartosci = W_horner(x_wartosci)

    with open("wielomian_wyniki_horner_vec.txt", "w") as file:
        for x, w in zip(x_wartosci, w_wartosci):
            file.write(f'W({x:.4f}) = {w:.4f}\n')
    
    end_time = time.time()
    print(f"Czas wykonania dziaania sposobem schematu Hornera: {end_time - start_time:.4f} sekundy")  


obliczenia_horner()
```
